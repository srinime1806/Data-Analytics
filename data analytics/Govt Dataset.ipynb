#  Crime Distribution by Category (Bar Chart)

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 5))
sns.barplot(x=df['HEADS CRIME'].value_counts().index, y=df['HEADS CRIME'].value_counts().values)
plt.xlabel("Crime Category")
plt.ylabel("Count")
plt.title("Crime Distribution by Category")
plt.xticks(rotation=45)
plt.show()
```

# Crime Trends Over Time

```python
import pandas as pd

df['SL.NO.'] = pd.to_numeric(df['SL.NO.'])  # Ensure it's numeric
df_sorted = df.sort_values(by='SL.NO.')

plt.figure(figsize=(10, 5))
plt.plot(df_sorted['SL.NO.'], df_sorted['During the current month'], label="Current Month")
plt.plot(df_sorted['SL.NO.'], df_sorted['During the previous month'], label="Previous Month", linestyle="dashed")
plt.xlabel("Record Number")
plt.ylabel("Crime Count")
plt.title("Crime Trends Over Time")
plt.legend()
plt.show()
```

# Finding Most Affected Crime Categories

```python
df.groupby('HEADS CRIME')['During the current month'].sum().sort_values(ascending=False).head(10)
```

# Predict Future Crime Trends

```python
df['Rolling_Avg'] = df['During the current month'].rolling(window=5).mean()

plt.figure(figsize=(10, 5))
plt.plot(df['SL.NO.'], df['During the current month'], label="Actual Crime", alpha=0.5)
plt.plot(df['SL.NO.'], df['Rolling_Avg'], label="Rolling Average (5 months)", linewidth=2)
plt.xlabel("Record Number")
plt.ylabel("Crime Count")
plt.title("Crime Trend Prediction using Rolling Average")
plt.legend()
plt.show()
```

#  Crime Rate Change (Current vs Previous Month)

# Independent T-Test 

```python
from scipy import stats

t_stat, p_value = stats.ttest_rel(df["During the current month"], df["During the previous month"])

print(f"T-Statistic: {t_stat}")
print(f"P-Value: {p_value}")

if p_value < 0.05:
    print("Significant difference: Crime rate changed between the two months.")
else:
    print("No significant difference: Crime rate remained stable.")
```
